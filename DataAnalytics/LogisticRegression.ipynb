#**Extracting Independent and Dependent Variable**

In the code below, we are going to **save our independent variable** into the list of **feature_cols**

Remember, our IndVariables are: 

'Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age',

Then, we are going to save these info into **variable X**

```python
feature_cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 
                'DiabetesPedigreeFunction', 'Age']

X = df[feature_cols]                        #These are Independent Variables
y = df.Outcome                              #This is the Dependent Variable/Target
```

#**Splitting dataset into training and testing dataset**

Remember what we learned before,

Training set is for **model training**

Testing set is for **predictions and evaluate model performance**

##**Importing packages**

```python
from sklearn.model_selection import train_test_split
```

##**Testing & Training**

The code below splits the dataset

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

##**Model development and Predictions**

This time, we are using **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

                  ## Creating a Logistic Regression Classifier Object ##

logreg = LogisticRegression()

                           ## Fitting the model with data ##

logreg.fit(x_train,y_train)

```

**Once the model is fitted, now let's print our predictions!**

```python
                           ## Making predictions ##

y_pred = logreg.predict(x_test)
y_pred
```

#**Evaluation of model using Confusion Matrix**

Confusion matrix --> is a table that is often used to describe the performance of a classification model (or a classifier)

Relatively simple to understand but terminology can be confusing

Basically, **confusion matrix** shows the ways in which **our classification model is confused** when it **makes predictions on observations**

It helps us to measure the type of error our model is making & classifying the observation into different classes

Simple terms -- **It is a matrix used to evaluate the performance of classification model**

##**Key parts of Confusion Matrix**

1. True Positive (TP)
--> Cases in which we predicted **YES** and prediction actually **TRUE**

--> E.g : Person **IS DIABETIC** and we predicted it to be **TRUE**

2. True Negative (TN)
--> Cases which we predicted **NO** and prediction actually **TRUE**

--> E.g : Person **IS NOT DIABETIC** and our prediction **TRUE** 

--> Basically means that we predict the person is not diabetic, and turns out he wasn't so our prediction is true

3. False Positive (FP)
--> Cases which we predicted **YES** but our prediction is **FALSE**

--> E.g : Person **IS NOT DIABETIC** but our model predicted that **he/she IS DIABETIC**

4. False Negative (FN)

--> Refers to cases in which we predicted **NO** but our prediction turned out to **FALSE**

--> E.g : Person **IS DIABETIC** but our model predicted that **he isn't**


##**More example of Confusion Matrix**

Confusion matrix will give you a matrix, and using that matrix we can understand where our model **gets confused while making predictions**

matrix.JPG

The classifier made a total of **165 predictions**

Out of 165 predictions, classifier predicted **"YES" 110 times**, and **"NO" 55 times**

In the table above things that you don't need to **worry about/our prediction is correct**:

1. Actual NO + Predicted NO

--> In actual, the person **no diabetes**, and our prediction also say that he/she **no diabetes**

2. Actual YES + Predicted YES

--> The person **has diabetes**, and our prediction also say that he/she **has diabetes**


Things that **YOU SHOULD WORRY ABOUT** is 

1. Actual NO + Predicted YES

--> The person is **fit & healthy**, but our model predicted that the person **has diabetes**

2. Actual YES + Predicted NO

--> The person **has diabetes**, but our model predicted that the person **is not diabetic**
