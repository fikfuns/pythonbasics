#**Importing 'datetime' module**

```python
from datetime import datetime
```

##**Creating a date using year, month, day as arguments**

datetime(year, month, day[, hour[, minute[, second[, microsecond[,tzinfo]]]]])

The year, month and day arguments are required. tzinfo may be None, or an
instance of a tzinfo subclass. The remaining arguments may be ints.

```python
birthday = datetime(2002,1,11, 7, 30, 51)
birthday.year
#birthday.month
#birthday.day
```

0 is monday 6 is sunday

```python
birthday.weekday()
```

##**Creating a date using datetime.now()**

```python
datetime.now()

#Subtracting datetime

datetime(2018,1,1) - datetime(2017,1,1)
datetime.now() - datetime(2019,12,30)
```

##**Parsing a date using strptime**

string parse time, alot of times user enter **dates as string**

[datetime documentation](https://docs.python.org/3/library/datetime.html)

```python
parsed_date = datetime.strptime('Dec 12, 2019','%b %d, %Y')
parsed_date.month
```

```python
parsed_date.year
```

```python
parsed_date.day
```

##**Rendering a date as a string using strftime**

grab specific parts of date and convert it into a string

strftime

string format time

```python
date_string = datetime.strftime(datetime.now(), '%b %d, %Y')
date_string
```
