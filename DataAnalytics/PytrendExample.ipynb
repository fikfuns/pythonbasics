```python
import pandas as pd
from pytrends.request import TrendReq
pytrend = TrendReq()
```

#**Gathering Data from Google**

In this example we will be gathering data about Covid-19 Vaccination

vv Specify what are you looking for? vv

```python
pytrend.build_payload(kw_list=['covid','vaccination','vaccine','does covid vaccine work'])
```

2 ways of performing Search operation

**area of interest = Malaysia**

```python
# Google trends showing Interest by Region
# Interest in Worldwide vv

worlddata = pytrend.interest_by_region() # This will fetch worldwide information
worlddata
```

**max_rows** unrestricts the no. of rows displayed

#**Displaying the data in graphs**

```python
# Plotting the result on bar graph

worlddata.reset_index().plot(x='geoName',y='covid',figsize=(120,10),kind='bar') #figsize(width,height)
```

##**Gather data related to our query**

```python
related_queries = pytrend.related_queries()
```

```python
related_queries.values()
#rising --> people start looking for the keywords aswell
```

##**Analysing data related to Malaysia**

```python
kw_list = ['vaccine covid','vaccination covid','covid 19 vaccine']
pytrend.build_payload(kw_list,timeframe='2020-01-01 2020-12-28',geo='MY')

# timeframe --> specify a time limit, date to start from and end from
# YYYY-MM-DD
# geo --> two letters abbreviation of countries MY = Malaysia
```

##**Showing data based on Interest Over Time**

```python
malaysia_data = pytrend.interest_over_time()
malaysia_data

#isPartial --> whether google has saved this data / data is not complete if False
```

##**Displaying data in graphical visualisation**

```python
import matplotlib.pyplot as plt 
malaysia_data = malaysia_data.drop(labels=['isPartial'],axis = 'columns') #Delete 'columns' named 'isPartial'
image = malaysia_data.plot(title='Are Malaysians aware of Vaccines?')
fig = image.get_figure()

#Modifying the graph
#set_size_inches(width,height)
fig.set_size_inches(18.5,10.5)
```

###**Saving the report in a csv file format**

```python
malaysia_data.to_csv('covidvaccine_searches.csv')
```
