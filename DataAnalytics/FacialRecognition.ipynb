###**Installing CV & setting objects**

```python
import cv2
from google.colab.patches import cv2_imshow

# Specify the location of image and HaarCascade files

cascadeFile = 'haarcascade_frontalface_default.xml'
imagePath = 'akshay.jpg'

# Creating the HaarCascade Object

faceCascade = cv2.CascadeClassifier(cascadeFile)
```

##**Read the image(still images)**

cv2.imread()

image read

Whenever you **process an image**, convert it into a grayscale (it makes the process a bit faster) -- For processing only

IF you use coloured images, it's going to take awhile

```python
image = cv2.imread(imagePath)
```

###**Converting image to grayscale**

BGR - blue green red

bgr is a standard color format for pictures taken by a camera

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

##**Detecting face in an image**

method used : detectMultiScale(parameters) --> used to detect objects

[documentation on parameters used(read this its damn important)]((https://stackoverflow.com/questions/36218385/parameters-of-detectmultiscale-in-opencv-using-python))

**1st parameter passed in --> grayScale image (var = gray,in this code)**

**2nd parameter --> scaleFactor (How much the image size is reduced at each image scale?)**

scaleFactor allows the machine will to detect images in various sizes of 

*   scaleFactor allows the machine will to detect images in various sizes of 
images, small/big


**3rd parameter --> minNeighbors**

minNeighbors will affect the quality of detected faces

1.   When our machine detects a face, it will draw rectangles around it
2.   Sometimes it may draw more than 1 rectangles, so we need to specify a value, preferably 3~6


**4th parameter - minSize(width,height)**

*   This to specify minimum possible object size
*   Objects that are smaller will be ignored (smaller than W,H is ignored)
*   How small is the size the face you want to detect?







**The numbers that you receive is used to draw a rectangle around the face (output of faces)**

###**Drawing a rectangle around the face**

using cv2.rectangle(originalImage, startPoint, endPoint, colorcode(inRGB), thickness(thiccnessOfRectangle)

endpoints.png

starting point is X,Y,M,H



**Code explained**

cv2.rectangle(originalimage, (startpoint1,spoint2), (endpoint(referpic), (endpoint2), (color rgb code), (thiccness)

**Displaying final output**

```python
faces = faceCascade.detectMultiScale(gray, scaleFactor=1.5,
                                     minNeighbors=3,
                                     minSize=(30,30))

for x,y,w,h in faces:
  cv2.rectangle(image , (x,y), (x+w , y+h) , (0,255,0), 2)



cv2_imshow(image)
```

##**Running same algorithm on caprice.png**

```python
cascadeFile = 'haarcascade_frontalface_default.xml'
imagePath = 'caprice.png'

# Creating the HaarCascade Object

faceCascade = cv2.CascadeClassifier(cascadeFile)
image = cv2.imread(imagePath)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

faces = faceCascade.detectMultiScale(gray, scaleFactor=1.5,
                                     minNeighbors=3,
                                     minSize=(30,30))

for x,y,w,h in faces:
  cv2.rectangle(image , (x,y), (x+w , y+h) , (0,255,0), 2)



cv2_imshow(image)
```
