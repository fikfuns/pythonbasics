#**Understanding Pandas**


```python
import numpy as np

arr = np.array([10,20,30,40])
print(arr)
print(type(arr))
```

```python
import pandas as pd
data = pd.Series(arr)
print(data)

#To print data without labels,

print(data.values)
```

###**Series**


####**Storing data into series**


```python
myData = pd.Series(data=[23,45,18], index=['firstPersonAge','secondPersonAge',
                                           'thirdPersonAge'])
print(myData)
print(myData['firstPersonAge'])
print(myData[0])
```

```python
#For loops in pandas

for everyAge in myData:
  print(everyAge)

#But how to use for loop to print the names/index?

for age in myData.iteritems():
  print(age)
```

###**DataFrames**

```python
#Traditionally dictionaries go like this..
stInfo = {'Id':[1,2,3], 'Name':['John','Smith','Alex'],
          'Course':['AI','DS','DA']}

print(stInfo)
```

Converting the data into a tabular fashion, it makes it easy to read

```python
data = pd.DataFrame(stInfo)
data  

#See the difference in the formatting 
#print(data) --> Uncomment to check!
```
