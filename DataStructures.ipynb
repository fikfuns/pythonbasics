#**Lists In Python**

#is a collection of data values
# denoted by using [..]
#list is mutable --> changeable/modifiable contents in a list
#allows user to have duplicate elements


```python
##Creating A List

mylist = [10,20,30,40,50,60,70]
print(type(mylist))
print(mylist)

mylist1 = list([100,200,300,400])
print(type(mylist1))

##Creating an empty list

mylist2 = []
```

```python
#Lists store information as indexes
#Indexing examples
mylist = [10, 20, 30, 40, 50, 60, 70]
print(mylist[0])
#negative indexing
print(mylist[-1])
#slicing
print(mylist[1:5]) #(list[inclusive:EXCLUSIVE]),,list[]
```

```python
# iteration over a list
mylist = [10, 20, 30, 40, 50, 60, 70]

for everyValue in mylist:
  print(everyValue)
```

```python
#Len function = len(list) determine how many elements in a list
mylist = [10, 20, 30, 40, 50, 60, 70]
print(len(mylist))
```

```python
#Examples of List with while loops

i = 0
while i < len(mylist):
  print(mylist[i])
  i = i + 1
```

```python
# Understanding mutable lists
mylist = [10, 20, 30, 40, 50, 60, 70]
print(mylist)

mylist[1] = 200
print(mylist)
```

```python
#Modifying multiple items
mylist = [10, 20, 30, 40, 50, 60, 70]
print(mylist)

mylist[:3] = [100, 200, 300]
print(mylist)

```

```python
# List built-in methods

mylist =  [10,20,30,40,50,60,70,80]
print(mylist)

#1. append(value) --> used to add an item at the end of a list
mylist.append(90)
print(mylist)

#2. insert (index,value) --> used to add an item at specific positions
mylist.insert(1,200)
print(mylist)

#3. pop(index) --> used to delete an item from the specified index
mylist.pop(1)
print(mylist)

#4. pop() --> if no argument specified, it will delete the last element
mylist.pop()
print(mylist)

# 5. remove(value)
mylist.remove(80)
print(mylist)

# assuming many duplicates..
list_10 = [10,20,30,40,10,10,50,60,70,10,10]
list_10.remove(10)  #it will always delete the first occurence
print(list_10)

# 6. reverse() --> used to print list in reverse order
mylist.reverse()
print(mylist)
```

# In which situation can we use sets?
All data structure is equally important.

Sets are useful when you want to find out the **UNIQUE VALUES** from a dataset/list/tuple




Example : Write a program that should print number of unique values in **list**.

```python
mylist = [10,20,30,10,30,40,50,60,30,70]
#Use for loop to iterate mylist to count unique elements.

empty_list = []

for i in mylist:              
  if i not in empty_list:
    empty_list.append(i)

print("Number of unique values are:",len(empty_list))
```

**Step-by-step iteration of the code above**

1st iteration;;

 i == 10, if **10 does not exist in empty_list** == **False**;;

 5th iteration;;

 empty_list = [10,20,30]

 empty_list = [10,20,30,40,50]

 8th iteration;;
 i == 60, if **60 does not exist in empty_list** == **True**;;
 empty_list = [10,20,30,40,50,60]

# **Dictionaries**

**Used to store information in key:value format**

**denoted by using {}**

normal data storage,

stInfo = [123, "Ali" , 23 , "Male" , "India"]

             0    1    2     3       4

print(stInfo[2]) ==> 23

print(stInfo[4]) ==> India

The problem is : **What if we want to store lots of data?** **It's easier to remember names than numbers.**

**Dictionaries help us by assigning keys to an element** Example..

stInfo = {123 , "Ali", 23, "Male", "India"}
           
            rno   name age gender country


**How to get country information?**

print(stInfo['country'] ==> India

**Creating a dictionary**

variable = {key:value} 

**key is always a STR value** sometimes numbers but str>numbers

**value can be anything**


# **dictionaries are immutable, keys must be unique**

**Do not repeat your keys!!**



```python
stInfo = {'rno' :123, 'name':'Ali', 'age':23}
print(type(stInfo))
print(stInfo)

print(stInfo['rno'])
print(stInfo['name'])
print(stInfo.get('age'))


```

**What if we have same keys in 1 dict?**

stInfo = {'rno' :123, 'name' : 'Ali', 'age' : 23, 'rno': 456}

print(stInfo) will output..

**{'rno': 456**, 'name' :'Ali', 'age': 23}

 Data for 123 will be lost

# **Understanding Iteration in Dictionaries**

normal **for** loops will always **print keys information for user**


```python
for keys in stInfo:
  print(keys)
```

**Can we print dictionary values using for loops?**

for <temp.variable> in dictionary.values()..

```python
for val in stInfo.values():
  print(val)
```

**Can we print keys & values together?**

declare 2 variables, **one for keys one for values**

for **keysVariable,valuesVariable** in **dictionary.items()**:

  print(**keysVariable**,**valuesVariable**)

```python
for k,v in stInfo.items():
  print(k,v)

#For extra style points

for k,v in stInfo.items():
  print(k,'-->',v)
```
