```python
x = [5,8,10]
y = [12,16,6]
plt.plot(x,y)
plt.title('Info')
plt.ylabel('This is the Y axis')
plt.xlabel('This is the X axis')
plt.show()
```

##**Adding styles to graphs**

 

[Many more styles here](https://matplotlib.org/stable/tutorials/introductory/customizing.html)

[Wtf is ggplot?](https://github.com/yhat/ggpy)




```python
from matplotlib import style
style.use('ggplot')
x = [5,8,10]
y = [12,16,6]
x2 = [6,9,11]
y2 = [6,15,7]
plt.plot(x,y,'g', label='Line One', linewidth=5) #(x,y,'InsertColorCodeHere')
plt.plot(x2,y2,'c', label='Line Two', linewidth=3)
plt.title('Info')
plt.ylabel('This is the Y axis')
plt.xlabel('This is the X axis')
plt.legend() #This will display our labels @Line 7/Line 8
plt.grid(True,color='k') # k --> black colour (See those black line grids? (background))
plt.show()
```

#**Generating Scatter plots**

--> used in a **huge** values/size of data

--> show distribution of data

-->scatter(YourXAxis,YourYAxis,label='LabelNameHere', color ='SomeCodesofColor', s=SizeNumber, marker = 'InsertMarkerType'

[Marker Types](https://matplotlib.org/stable/api/markers_api.html)

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,1,3,9,2,6]
plt.scatter(x,y,label='scatter',color='g',s=25,marker='o')
plt.title('This is a Scatter Graph')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')
plt.legend()
plt.show()
```

#**Creating Pie Plots**

--> Data but in pizza form

```python
slices = [7,2,2,13]
activities = ['Sleep','Eating','Playing','Working']
cols = ['c','m','r','y'] #--> color = ['ColorCode1','ColorCode2','ColorCode3',..etc]

plt.pie(slices,               # your slices
        labels = activities,  # Labels obviously
        colors = cols,        # Give em colors (that you set in cols)
        startangle = 90,      # Straightens the angle in the pie chart
        shadow = True,        # Cast shadows in pie chart so it looks cool
        explode = (0,0.1,0,0), #Since eating is 0.1 the pie is going out abit
        autopct = '%1.1f%%')   # Overlay percentage over graph 
plt.show()
```

#**Changing Location of Legends**

[location codes to pass in plt.legend(loc=?)](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html)

by default the loc=0, which sets it automatically as best position

'best' = 0

'upper right'=1

'upper left'=2

'lower left'=3

'lower right'=4

'right'=5

'center left'=6

'center right'=7

'lower center'=8

'upper center'=9

'center'=10

```python
style.use('ggplot')
x = [5,8,10]
y = [12,16,6]
x2 = [6,9,11]
y2 = [6,15,7]
plt.plot(x,y,'g', label='Line One', linewidth=5) #(x,y,'InsertColorCodeHere')
plt.plot(x2,y2,'c', label='Line Two', linewidth=3)
plt.title('Info')
plt.ylabel('This is the Y axis')
plt.xlabel('This is the X axis')
plt.legend(loc=3) 
plt.grid(True,color='k') # k --> black colour (See those black line grids? (background))
plt.show()
```
